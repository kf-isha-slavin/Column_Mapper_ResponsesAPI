```python
import numpy as np 
import pandas as pd 
import math 
import re
```

```python
KF_template = pd.read_excel('./../Files/KF/ORGDC_BA_General_en_IDC 1.xlsx', sheet_name = 'Employee Data Requirements')
print(KF_template.shape)
KF_template.head(10)
```

```python
# Step 2: Extract the relevant rows
descriptions = KF_template.iloc[5]  # Row 8 (0-based index)
statuses = KF_template.iloc[6]      # Row 7 (0-based index)
titles = KF_template.iloc[8]        # Row 10 (0-based index)

print(descriptions.head(5))
print('\n')
print(statuses.head(5))
print('\n')
print(titles.head(5))
```

```python
# Step 3: Create the new DataFrame
new_KF_df = pd.DataFrame({
    'Column_Title': titles,
    'Description': descriptions,
    'Status': statuses
})
```

```python
# remove '\n' from descriptions, statuses, and titles
new_KF_df['Description'] = new_KF_df['Description'].str.replace('\n', ' ', regex=False)
new_KF_df['Status'] = new_KF_df['Status'].str.replace('\n', ' ', regex=False)
new_KF_df['Column_Title'] = new_KF_df['Column_Title'].str.replace('\n', ' ', regex=False)

new_KF_df.head(10)
```

```python
# reset index
new_KF_df = new_KF_df.reset_index(drop=True)
```
